## Single Linear Regression

```python
import pandas as pd
data=pd.read_csv("Data.csv")
data
```

```python
X=data.iloc[:,0:1]
Y=data.iloc[:,-1]
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

lr.fit(X,Y)
Y_pred=lr.predict(X)
print(Y_pred)
```

```python
diff=pd.DataFrame({"Actual":Y,"Predicted":Y_pred})
diff
```

```python
print("Coefficient:",lr.coef_)
print("Intercept:",lr.intercept_)
```

```python
from sklearn import metrics

print("MAE:",metrics.mean_absolute_error(Y,Y_pred))
print("MSE:",metrics.mean_squared_error(Y,Y_pred))
print("R2 score:",metrics.r2_score(Y,Y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y,color='blue',label="Actual Value")
plt.plot(X,Y_pred,color='red',label="Predicted Value")
plt.legend()
plt.show()
```

## Plynomial Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd 

pol=pd.read_csv("polylinearregression.csv")
pol
```

```python
x=pol.iloc[:,1:2]
y=pol.iloc[:,2]

poly=PolynomialFeatures(degree=2)
model_poly=poly.fit_transform(x)
lr=LinearRegression()
lr.fit(model_poly,y)
```

```python
y_pred=lr.predict(model_poly)
print(y_pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.scatter(x,y,color='blue')
plt.plot(x,y_pred,color='red')
plt.xlabel("Temperature")
plt.ylabel("Pressure")
plt.title("Polynomial Regression")
plt.show()
```

```python
from sklearn import metrics
print("MSE:",metrics.mean_squared_error(y,y_pred))
print("R2 Score:",r2_score(y,y_pred))
print("RMSE:",np.sqrt(metrics.mean_absolute_error(y,y_pred)))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd

x=[1,2,3,4,5]
y=[1,4,9,16,25]
x=pd.DataFrame({"x":x})
y=pd.Series(y)

poly=PolynomialFeatures(degree=1)
model_poly=poly.fit_transform(x)
lr=LinearRegression()
lr.fit(model_poly,y)
```

```python
y_pred=lr.predict(model_poly)
print(y_pred)
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
print("MSE:",metrics.mean_squared_error(y,y_pred))
print("R2 Score:",r2_score(y,y_pred))
print("RMSE:",np.sqrt(metrics.mean_absolute_error(y,y_pred)))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y,color='blue')
plt.plot(x,y_pred,color='red')
plt.title("Polynomial Regression")
plt.show()
```
